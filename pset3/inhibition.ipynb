#### Somatic inhibition

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Gs = 1
Gc = 1 / (9 * Gs)
Gd = Gs
alpha = [0, 0.2, 0.5, 1, 2, 5]
Ee = 100 #mV
x = np.logspace(-2, 3, 10000)

plt.figure(0, figsize=(12, 8))
plt.axes(xscale = "log") 
for a in alpha:
    Gi = a * Gd
    Vs = np.zeros(x.shape)
    for i in range(10000):
        Vs[i] = (x[i]*Gc*Ee) / (Gc*Gd+Gs*Gd + Gs*Gc + x[i]*(Gc+Gi+Gs)+ Gi*(Gd+Gc))
    
    plt.plot(x, Vs, label='alpha = {}'.format(a))

plt.legend()
plt.xlabel("Excitory Conductance (G_e)", fontsize=20)
plt.ylabel("Membrane Potential at Soma (V_s)", fontsize=20)
plt.xticks([0.01, 0.1, 1, 10, 100, 1000])
plt.show()
```
