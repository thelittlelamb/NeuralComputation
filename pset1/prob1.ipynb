## Problem 1: Diffusion in 1D

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import loadmat
```

```python
randomwalk = loadmat("randomwalk.mat")
randomwalk = randomwalk['X']
print(type(randomwalk))
print(randomwalk.shape)
```

```python
## 1. the initial postion of all particles
init = randomwalk[:, 0]
# print(init) -> all particles start at position 0
## 2. the last postion of all particles
last = randomwalk[:, -1]
# print(last)
```

```python
## 3. the trajectory of particle 14
p14 = randomwalk[13]
time = np.arange(p14.shape[0])
## plt: syntax
# plot([x], y, [fmt], *, data=None, **kwargs)
# plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs)
plt.figure(1)
plt.plot(time, p14)
plt.title("The trajectory of particle 14")
plt.xlabel("Time")
plt.ylabel("Trajectory")
plt.show()
```

```python
## 4. The trajectory of all particles
plt.figure(2)
for i in range(randomwalk.shape[0]):
    plt.plot(randomwalk[i])
plt.title("The trajectory of all particles")
plt.xlabel("Time")
plt.ylabel("Trajectory")
plt.show()
    
```

```python
## 5. The mean displacement of particles (as a function of time)
# mean_of_each_particle = np.mean(randomwalk, axis = 1)
mean_of_all_particle = np.mean(randomwalk, axis = 0)
plt.figure(3)
plt.plot(mean_of_all_particle)
plt.title("The mean displacement")
plt.xlabel("Time")
plt.ylabel("Mean Displacement")
plt.show()
# Interpret: the average->around zero
```

```python
## 6 The mean square displacement
## delta = 1, time = 1 unit
distance_square = np.square(randomwalk)
mean_square_of_all_particle = np.mean(distance_square, axis = 0)
plt.figure(4)
plt.plot(mean_square_of_all_particle)
plt.title("The mean square displacement")
plt.xlabel("Time")
plt.ylabel("Mean Square Displacement")
plt.show()
## a linear relationship
```
